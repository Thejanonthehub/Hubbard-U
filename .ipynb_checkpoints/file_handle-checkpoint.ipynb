```python
# data extraction by formula


import os
import pandas as pd
from mp_api.client import MPRester
from pymatgen.core import Structure
from pymatgen.analysis.local_env import CrystalNN
from pymatgen.core.periodic_table import Element
from tqdm import tqdm

# ============================
# User settings
# ============================
API_KEY = api   # Replace with your Materials Project API key
input_csv = "hubbard_u_j_values.csv"   # CSV with columns ["formula", "Species"]
output_csv = "hubbard_features.csv"
cif_folder = "./hubbard_structures_cifs"  # Folder containing local CIF files named like "NaAgO.cif"

# Load dataset
df = pd.read_csv(input_csv)

# Prepare list to store features
features_list = []

# Connect to Materials Project
mpr = MPRester(API_KEY)

# CrystalNN for coordination/bond lengths
cnn = CrystalNN()

# Loop through each row with progress bar
for idx, row in tqdm(df.iterrows(), total=len(df), desc="Extracting features"):
    formula = row["formula"]
    species = row["Species"]
    structure = None

    # -----------------------------
    # 1. Try Materials Project first
    # -----------------------------
    try:
        docs = mpr.materials.summary.search(
            formula=formula,
            fields=["structure", "symmetry", "volume", "nsites"]
        )

        if docs:
            doc = docs[0]
            structure = doc.structure
            spg_num = doc.symmetry.number if doc.symmetry else None
            volume_per_atom = doc.volume / doc.nsites
        else:
            spg_num = None
            volume_per_atom = None

    except Exception as e:
        print(f"⚠️ MP search failed for {formula}: {e}")
        spg_num = None
        volume_per_atom = None

    # -----------------------------
    # 2. Fallback to local CIF
    # -----------------------------
    if structure is None:
        cif_path = os.path.join(cif_folder, f"{formula}.cif")
        if os.path.exists(cif_path):
            try:
                structure = Structure.from_file(cif_path)
                spg_num = structure.get_space_group_info()[1]  # number
                volume_per_atom = structure.volume / structure.num_sites
            except Exception as e:
                print(f"❌ Failed to read CIF for {formula}: {e}")
                continue
        else:
            print(f"⚠️ No MP entry or CIF file for {formula}, skipping...")
            continue

    # -----------------------------
    # 3. Compute coordination & bond lengths
    # -----------------------------
    coord_nums = []
    bond_lengths = []

    for i, site in enumerate(structure):
        if site.species_string == species:
            try:
                neighbors = cnn.get_nn_info(structure, i)
                coord_nums.append(len(neighbors))
                bond_lengths.extend([n["weight"] for n in neighbors])
            except:
                continue

    avg_cn = sum(coord_nums)/len(coord_nums) if coord_nums else None
    avg_bond_len = sum(bond_lengths)/len(bond_lengths) if bond_lengths else None

    # -----------------------------
    # 4. Electronegativity difference with neighbors
    # -----------------------------
    if coord_nums:
        try:
            el_species = Element(species).X
            en_diffs = []
            for i, site in enumerate(structure):
                if site.species_string == species:
                    neighbors = cnn.get_nn_info(structure, i)
                    for n in neighbors:
                        neigh_el = Element(n['site'].species_string).X
                        en_diffs.append(abs(el_species - neigh_el))
            avg_en_diff = sum(en_diffs)/len(en_diffs) if en_diffs else None
        except:
            avg_en_diff = None
    else:
        avg_en_diff = None

    # -----------------------------
    # 5. Number of d-electrons
    # -----------------------------
    try:
        d_count = sum([occ for (n, l, occ) in Element(species).full_electronic_structure if l == 'd'])
    except:
        d_count = None

    # -----------------------------
    # 6. Collect features
    # -----------------------------
    features_list.append({
        "formula": formula,
        "Species": species,
        "spacegroup_number": spg_num,
        "volume_per_atom": volume_per_atom,
        "avg_coordination_number": avg_cn,
        "avg_bond_length": avg_bond_len,
        "avg_en_diff": avg_en_diff,
        "num_d_electrons": d_count
    })

# Save features to CSV
features_df = pd.DataFrame(features_list)
features_df.to_csv(output_csv, index=False)

print(f"✅ Feature extraction completed. Saved to {output_csv}")
```

```python
import os
import pandas as pd
from mp_api.client import MPRester
from pymatgen.core import Structure
from pymatgen.core.composition import Composition
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.analysis.local_env import CrystalNN
from pymatgen.core.periodic_table import Element
from tqdm import tqdm

# ============================
# User settings
# ============================
API_KEY = api  # Replace with your Materials Project API key
input_csv = "hubbard_u_j_values.csv"   # Must have ["identifier", "formula", "Species"]
output_csv = "x3.csv"  
cif_folder = "./hubbard_structures_cifs"  # Fallback CIF folder

# Load dataset
df = pd.read_csv(input_csv)

# Prepare list to store features
features_list = []

# Connect to Materials Project
mpr = MPRester(API_KEY)

# CrystalNN for coordination/bond lengths
cnn = CrystalNN()

# -------------------------
# Helper function to count d and f electrons robustly
# -------------------------
def count_df_electrons(symbol: str) -> float:
    """Count total d + f electrons for a given element."""
    el = Element(symbol)
    full_es = el.full_electronic_structure
    df_count = 0.0
    for orb in full_es:
        try:
            if len(orb) == 3:
                n, shell, occ = orb  # shell is str like 'd'
                if shell in ['d', 'f']:
                    df_count += occ
        except (ValueError, TypeError):
            continue  # Skip malformed entries
    return df_count

# -------------------------
# Helper function to get valence electrons
# -------------------------
def get_valence_electrons(symbol: str) -> float:
    """Return total number of valence electrons for transition metals (ns + (n-1)d)."""
    el = Element(symbol)
    full_es = el.full_electronic_structure
    if el.block != 'd':
        # Fallback for non-d block: sum outermost shell
        max_n = max(n for n, shell, occ in full_es)
        return sum(occ for n, shell, occ in full_es if n == max_n)
    
    # For d-block: ns + (n-1)d
    row = el.row
    s_count = sum(occ for n, shell, occ in full_es if n == row and shell == 's')
    d_count = sum(occ for n, shell, occ in full_es if n == row - 1 and shell == 'd')
    return s_count + d_count

# ============================
# Main feature extraction loop
# ============================
for idx, row in tqdm(df.iterrows(), total=len(df), desc="Extracting features"):
    identifier = str(row["identifier"])
    formula = row["formula"]
    species = row["Species"]
    structure = None
    spg_num, volume_per_atom = None, None

    # -----------------------------
    # 1. Try Materials Project by identifier
    # -----------------------------
    try:
        if identifier.startswith("mp-"):
            docs = mpr.materials.summary.search(
                material_ids=[identifier],
                fields=["structure", "symmetry", "volume", "nsites"]
            )
            if docs:
                doc = docs[0]
                structure = doc.structure
                volume_per_atom = doc.volume / doc.nsites
    except Exception as e:
        print(f"⚠️ MP lookup failed for {identifier}: {e}")

    # -----------------------------
    # 2. Fallback to local CIF
    # -----------------------------
    if structure is None:
        cif_path = os.path.join(cif_folder, f"{identifier}.cif")
        if not os.path.exists(cif_path):
            cif_path = os.path.join(cif_folder, f"{formula}.cif")
        if os.path.exists(cif_path):
            try:
                structure = Structure.from_file(cif_path)
                volume_per_atom = structure.volume / structure.num_sites
            except Exception as e:
                print(f"❌ Failed to read CIF for {identifier} ({formula}): {e}")
                continue
        else:
            print(f"⚠️ No structure found for {identifier} ({formula}), skipping...")
            continue

    # -----------------------------
    # 3. Spacegroup, Bravais, lattice parameters
    # -----------------------------
    try:
        sga = SpacegroupAnalyzer(structure, symprec=0.01)
        spg_symbol = sga.get_space_group_symbol()
        spg_num = sga.get_space_group_number()
        bravais = sga.get_lattice_type()
        lattice = structure.lattice
        a, b, c = lattice.a, lattice.b, lattice.c
        alpha, beta, gamma = lattice.alpha, lattice.beta, lattice.gamma
    except Exception as e:
        print(f"⚠️ Symmetry analysis failed for {identifier}: {e}")
        spg_symbol = spg_num = bravais = None
        a = b = c = alpha = beta = gamma = None

    # -----------------------------
    # 4. Coordination & bond lengths
    # -----------------------------
    coord_nums, bond_lengths = [], []
    for i, site in enumerate(structure):
        if site.species_string == species:
            try:
                neighbors = cnn.get_nn_info(structure, i)
                coord_nums.append(len(neighbors))
                bond_lengths.extend([n["weight"] for n in neighbors])
            except Exception:
                continue

    avg_cn = sum(coord_nums)/len(coord_nums) if coord_nums else None
    avg_bond_len = sum(bond_lengths)/len(bond_lengths) if bond_lengths else None

    # -----------------------------
    # 5. Electronegativity difference
    # -----------------------------
    avg_en_diff = None
    if coord_nums:
        try:
            el_species = Element(species).X
            en_diffs = []
            for i, site in enumerate(structure):
                if site.species_string == species:
                    neighbors = cnn.get_nn_info(structure, i)
                    for n in neighbors:
                        neigh_el = Element(n['site'].species_string).X
                        en_diffs.append(abs(el_species - neigh_el))
            avg_en_diff = sum(en_diffs)/len(en_diffs) if en_diffs else None
        except Exception:
            pass

    # -----------------------------
    # 6. d/f electrons and valence
    # -----------------------------
    try:
        num_df_electrons_species = count_df_electrons(species)
        valence_species = get_valence_electrons(species)

        comp = Composition(formula)
        num_df_electrons_formula = sum(count_df_electrons(el_sym) * amt 
                                       for el_sym, amt in comp.element_composition.items())
    except Exception as e:
        print(f"⚠️ Electron count failed for {species} in {formula}: {e}")
        num_df_electrons_species = num_df_electrons_formula = valence_species = None

    # -----------------------------
    # 7. Collect features
    # -----------------------------
    features_list.append({
        "identifier": identifier,
        "formula": formula,
        "Species": species,
        "spacegroup_number": spg_num,
        "spacegroup_symbol": spg_symbol,
        "volume_per_atom": volume_per_atom,
        "bravais_lattice": bravais,
        "lattice_a": a,
        "lattice_b": b,
        "lattice_c": c,
        "lattice_alpha": alpha,
        "lattice_beta": beta,
        "lattice_gamma": gamma,
        "avg_coordination_number": avg_cn,
        "avg_bond_length": avg_bond_len,
        "avg_en_diff": avg_en_diff,
        "num_df_electrons_species": num_df_electrons_species,
        "num_valence_electrons_species": valence_species,
        "num_df_electrons_formula": num_df_electrons_formula
    })

# Save features to CSV
features_df = pd.DataFrame(features_list)
features_df.to_csv(output_csv, index=False)

print(f"✅ Feature extraction completed. Saved to {output_csv}")
```

```python
from pymatgen.core.periodic_table import Element

def get_valence_electrons(element_symbol: str):
    """
    Returns the number of valence electrons for a given element.
    Valence electrons = all electrons in the highest principal quantum number (n).
    """
    el = Element(element_symbol)
    full_es = el.full_electronic_structure  # [(n, shell, occ), ...]
    
    # Find the outermost (maximum n)
    max_n = max(n for (n, shell, occ) in full_es)
    
    # Count all electrons in the outermost shell
    valence_electrons = sum(occ for (n, shell, occ) in full_es if n == max_n)
    
    return valence_electrons

# Example: Silver (Ag)
el = "Ag"
valence = get_valence_electrons(el)
print(f"Element: {el}, Valence electrons = {valence}")

# Test a few more
for el in ["V"]:
    print(f"{el}: {get_valence_electrons(el)}")
```

```python
import os
from pymatgen.core import Structure
from pymatgen.analysis.local_env import CrystalNN
import pandas as pd

# Folder containing CIF files
cif_folder = "./hubbard_structures_cifs"

# Initialize CrystalNN (robust method for coordination analysis)
cnn = CrystalNN()

# List to store results
cn_data = []

# Loop through all CIF files in the folder
for cif_file in os.listdir(cif_folder):
    if cif_file.endswith(".cif"):
        cif_path = os.path.join(cif_folder, cif_file)

        try:
            structure = Structure.from_file(cif_path)

            # Compute CN for each site
            for i, site in enumerate(structure):
                cn = cnn.get_cn(structure, i)
                cn_data.append({
                    "file": cif_file,
                    "site_index": i,
                    "element": site.species_string,
                    "coordination_number": cn
                })
        
        except Exception as e:
            print(f"⚠️ Failed to process {cif_file}: {e}")

# Save to CSV
df = pd.DataFrame(cn_data)
df.to_csv("coordination_numbers.csv", index=False)

print("✅ Coordination numbers saved to coordination_numbers.csv")
```

```python
from pymatgen.core.periodic_table import Element

def get_valence_electrons(symbol: str) -> int:
    """
    Return total number of valence electrons for a given element.
    Uses outermost shell (highest principal quantum number).
    """
    el = Element(symbol)
    # Get the maximum principal quantum number (outer shell)
    max_n = max([orb[0] for orb in el.full_electronic_structure])
    # Sum all electrons in that shell
    valence_electrons = sum([orb[2] for orb in el.full_electronic_structure if orb[0] == max_n])
    return valence_electrons

# -------------------------
# Example Usage
# -------------------------
elements = ["Ga"]
for e in elements:
    print(f"{e}: {get_valence_electrons(e)} valence electrons")
```
